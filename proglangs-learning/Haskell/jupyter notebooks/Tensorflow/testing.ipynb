```haskell
{-# LANGUAGE ScopedTypeVariables #-}

import TensorFlow.Core
import TensorFlow.Logging
import TensorFlow.Ops
import TensorFlow.Types
```

```haskell
let (n1 :: Tensor Build Bool) = constant (Shape [1]) [True]
    (n2 :: Tensor Build Float) = constant (Shape [1]) [2.0 :: Float]
```

```haskell
:t n1
:t n2
```

```haskell
n1 `add` n2
```

```haskell
:t constant
let n3 = constant (Shape [1]) [1 :: Float]
    n4 = n2 `add` n3
```

```haskell
-- :t render n4
-- :t runSession . run $ render n4
-- :t readValue
-- :t n4
(l :: Scalar Float) <- runSession . run $ n4
print l
```

```haskell
let n5 = constant (Shape [2,3]) [56, 64, 324, 32, 6, 76 :: Float]
    n7 = n2 `add` n5 -- shouldn't work!, their shapes are different (or should it?)

(l :: Scalar Float) <- runSession . run $ n7
print l -- <- yep it fails!!!
```
